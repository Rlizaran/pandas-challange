```python
# Import librarys
import os
import pandas as pd
```

# Importing path to file from OS and open to read it

```python
# declare path for csv file and read it
cvspath = os.path.join('Resources','purchase_data.csv')
purchase_data = pd.read_csv(cvspath)
purchase_data
```

# Player Counts

```python
# Display total player count
player_grouped =purchase_data.groupby('SN')
players = len(player_grouped)
total_players = pd.DataFrame({'Total Players':[players]})
total_players
```

# Purchasing Analysis (Total)

```python
# Run basic calculations to obtain number of unique items, average price, etc.
item_name_group = purchase_data.groupby('Item Name')
number_of_Unique_Items = len(item_name_group)
number_of_purchases = len(purchase_data['Purchase ID'])
average_price = (sum(purchase_data['Price'])/len(purchase_data['Price']))
total_revenue = sum(purchase_data['Price'])
# Create a summary data frame to hold the results
summary = pd.DataFrame({'Number of Unique Items':[number_of_Unique_Items],'Average Price':average_price,'Number of Purchases':number_of_purchases,'Total Revenue':total_revenue})
# Optional: give the displayed data cleaner formatting
summary ['Average Price'] = summary ['Average Price'].map('${:.2f}'.format)
summary ['Total Revenue'] = summary ['Total Revenue'].map('${:.2f}'.format)
# Display the summary data frame
summary
```

# Gender Demographics

```python
# total number of Players
players

# Percentage and Count of Male Players
male_group = purchase_data.loc[purchase_data['Gender']=='Male','SN']
total_male = len(male_group.unique())
percentage_male = total_male / players*100

# Percentage and Count of Female Players
female_group = purchase_data.loc[purchase_data['Gender']=='Female','SN']
total_female = len(female_group.unique())
percentage_female = total_female/players*100

# Percentage and Count of Other / Non-Disclose
other_group = purchase_data.loc[purchase_data['Gender']=='Other / Non-Disclosed','SN']
total_other = len(other_group.unique())
percentage_other = total_other/players *100
```

```python
# Create Gender Demographics Data Frame and display a clean summary
gender_demographics = pd.DataFrame({'':['Male','Female', 'Other / Non-Disclosed'],'Total Count':[total_male,total_female,total_other],
                                    'Percentage of Players':[percentage_male,percentage_female,percentage_other]})
gender_demographics['Percentage of Players'] = gender_demographics['Percentage of Players'].map("{:.2f}%".format)
gender_demographics
```

#  Purchasing Analysis (Gender)

```python
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
group_gender = purchase_data.groupby(['Gender'])

# Total Purchase Value per Gender
total_purchase_per_gender = group_gender['Price'].sum()

# Total Purchase Count
total_purchase_count = group_gender['Purchase ID'].count()

#Average Purchase Price
average_price_per_gender = group_gender['Price'].sum()/group_gender['Purchase ID'].count()

# Avg Total Purchase per Person
avg_purchase_male = total_purchase_per_gender['Male']/total_male
avg_purchase_female =total_purchase_per_gender['Female']/ total_female
avg_purchase_other = total_purchase_per_gender['Other / Non-Disclosed']/total_other

# Create a summary data frame to hold the results
summary_raw = pd.DataFrame([total_purchase_count,average_price_per_gender,total_purchase_per_gender])
summary_raw = pd.DataFrame.transpose(summary_raw)
summary = summary_raw.rename(columns={'Unnamed 0':"Average Purchase Price",'Purchase ID':'Purchase Count','Price':'Total Purchase Value'})

# Add last column for Avg total purchase per person
summary['Avg Total Purchase per Person']=[avg_purchase_female,avg_purchase_male,avg_purchase_other]

# Optional: give the displayed data cleaner formatting
summary['Purchase Count'] = summary['Purchase Count'].astype(int)
summary['Average Purchase Price'] = summary['Average Purchase Price'].map("${:.2f}".format)
summary['Total Purchase Value'] = summary['Total Purchase Value'].map("${:.2f}".format)
summary['Avg Total Purchase per Person'] = summary['Avg Total Purchase per Person'].map("${:.2f}".format)

# Display the summary data frame
summary
```
